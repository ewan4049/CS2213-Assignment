## World Happiness Report (2023)
The World Happiness Report (2023) contains data ranking countries based on their happiness scores, derived from various economic, social, and personal factors.



````python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/andvise/repo/refs/heads/main/happiness.csv")


df.dtypes
````

* Country name: Name of the country.

* Ladder score: Overall happiness score (higher is happier).

* Logged GDP per capita: Economic strength measured by GDP per person (logged).

* Social support: Level of social support experienced by individuals.

* Healthy life expectancy: Average healthy lifespan of individuals.

* Freedom to make life choices: Perceived freedom individuals have in life decisions.

* Generosity: Measure of generosity among the population.

* Perceptions of corruption: Public perception regarding corruption in the government/business.

* Region: Region to which the country belongs.

````python
df[["Country name", "Ladder score", "Logged GDP per capita", "Region", "Healthy life expectancy"]]
````

#### Q2: Are richer countries (GDP per Capita) generally happier?


````python
df.plot.scatter(x = "Logged GDP per capita", y = "Ladder score", title = "Are richer countries (GDP per Capita) generally happier?", c = "green")
````

While there are certainly outliers, there is generally speaking a positive correlation between GDP and happiness, with all of the highest scoring happiness scores belonging to richer countries.

#### Q3: Which regions are more represented in the dataset? What percentage of the total number of countries do they represent?


````python
df_region = df.groupby("Region")["Region"].count().plot.pie(ylabel = "", autopct='%1.1f%%', figsize=(6, 6), legend = True, labeldistance = None).legend(loc=(1.2,0.7))
````

The largest regions represented in the graph are Subsaharan African, Western Europe and Central and Eastern Europe. Subsaharan Africa represents 23.4% of the data, while Europe's 12.4% and 14.6% scores represent a combined 27% of the data.

#### Q4: What is the most common happiness score? Is it more common to have high or low levels of happiness?

````python
df["Ladder score"].plot.hist(title = "Frequency of happiness ladder scores", xlabel = "Happiness scores", ylabel = "Frequency", bins = 70, color = "purple")
````

Here, we can see that the most common happiness score is around 6 on the ladder score. Additionally, we can see that higher levels of happiness are more common, as there is significantly more data on the higher, right side of the graph.

#### Q5: Which region has a lower healthy life expectancy?

````python
df.plot.box(column = 'Healthy life expectancy', by = 'Region', fontsize = 10, rot = 90, xlabel = "Countries", ylabel = "Average life expectancy")
````

From this, we can see that the Middle East and North Africa have the lowest life expectancy scores.
